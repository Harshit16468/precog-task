```python
import nltk
from nltk.corpus import wordnet
from nltk.corpus import reuters

from nltk.corpus import brown


reuters_words = brown.words()

distinct_words = set()

for word in reuters_words:
    synsets = wordnet.synsets(word)
    if len(synsets) > 1:
        similar_word_exists = any(w in distinct_words for w in synsets[0].lemma_names())
        if not similar_word_exists:
            distinct_words.add(word)

with open('distinct.txt', 'w') as file:
    for word in distinct_words:
        file.write(word + '\n')
```

```python
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from collections import Counter
import math
from nltk.corpus import reuters
from nltk.tokenize import word_tokenize
from math import sqrt
import pandas as pd
from scipy.stats import spearmanr
from sklearn.decomposition import PCA
import pandas as pd
from tqdm import tqdm
from scipy.stats import spearmanr
import numpy as np

simlex_subset = pd.read_csv('/home/harshit/Desktop/random/precog/SimLex-999/SimLex-999.txt', sep='\t')

word1_list = simlex_subset['word1'].tolist()
word2_list = simlex_subset['word2'].tolist()
```

```python
with open('distinct.txt', 'r') as file:
    distinct_loaded_words = [line.strip() for line in file]
predictions=[]
```

```python
def get_cosine(vector1, vector2):
    intersection = min(len(vector1), len(vector2))
    numerator = sum(vector1[i] * vector2[i] for i in range(intersection))
    sum1 = sum(vector1[i] ** 2 for i in range(len(vector1)))
    sum2 = sum(vector2[i] ** 2 for i in range(len(vector2)))
    denominator = sqrt(sum1) * sqrt(sum2)
    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator
def text_to_vector(text):
    vector1=[]
    synsets1 = wordnet.synsets(text)
    for word in (distinct_loaded_words):
        flag=0
        synsets2 = wordnet.synsets(word)
        for synset1 in synsets1:
            for synset2 in synsets2:
                similarity = synset1.wup_similarity(synset2)
                if similarity is not None:
                    vector1.append(similarity)
                    flag=1
                    break
            if flag==1:
                break
        
        if flag==0:
            vector1.append(0)
    return vector1
```

```python
for i in tqdm(range(0,len(word1_list))):
    text1 = word1_list[i]
    text2 = word2_list[i]
    vector1 = text_to_vector(text1)
    vector2 = text_to_vector(text2)
    cosine = get_cosine((vector1), (vector2))
    predictions.append(cosine)
ground_truth_scores = simlex_subset['SimLex999'].values
correlation, _ = spearmanr(predictions, ground_truth_scores)
print(f'Spearman Correlation: {correlation*100}%')
```
